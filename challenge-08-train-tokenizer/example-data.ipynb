## Example data for training tokenizer and model

A tiny, tiny, tiny sample of the text we use to train the tokenizer and base train the model.

We use a few billion characters of text to train our tokenizer.

We use something like 54 billion characters of text to train the d20 model.

We'll use around 180 billion characters of text to train the d32 model.

For comparison, according to [this article](https://en.wikipedia.org/wiki/Wikipedia:Size_of_Wikipedia), there are around 30 billion characters in the English version of Wikipedia.

```python
import sys
sys.path.append('../my_nanochat')
import random
import textwrap
from my_nanochat.my_dataset import parquets_iter_batched
```

### First 50 characters...

Here are the first 50 characters of a random sample of 0.1% of the texts from the first 10 shards.

(The reason it's only the first 10 shards is I downloaded 11 shards of data to my mac and the last one gets used for validation.)

```python
random.seed(1)
for i, texts in enumerate(parquets_iter_batched('train')):
    for j, text in enumerate(texts):
        if random.random() < 0.001:
            excerpt = text[:50].replace('\n','\\n')
            print(f"{i:04d},{j:04d}: {excerpt}")
```

### Full text...

Here's the full text a random sample of 0.002% of the texts from the first 10 shards.

```python
random.seed(5)
for i, texts in enumerate(parquets_iter_batched('train')):
    for j, text in enumerate(texts):
        if random.random() < 0.00002:
            print(f"======={i}, {j}=======\n{textwrap.fill(text, width=90, replace_whitespace=False)}\n")
```
