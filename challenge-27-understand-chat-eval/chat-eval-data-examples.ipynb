# Chat eval data examples

A few examples of each data type "task" used for chat eval. This is similar and somewhat overlapping with `challenge-26-understand-midtrain/midtrain-data-examples.ipynb` but there are some tasks used for mid training and not chat eval and vice versa. The point of this notebook is to show all tasks used for chat eval in one place.

```python
import sys
sys.path.append('../my_nanochat')
import os
from my_tasks.humaneval import HumanEval
from my_tasks.my_mmlu import MyMMLU
from my_tasks.my_arc import MyARC
from my_tasks.my_gsm8k import MyGSM8K
from my_tasks.my_spellingbee import MySpellingBee
```

```python
def print_conversation(conversation):
    for message in conversation['messages']:
        print(f"_____{message['role']}_____")
        content = message['content']
        if isinstance(content, str):
            print(content)
        else:
            assert isinstance(content, list)
            for part in content:
                print(f"{part['type']}: {part['text']}")
        print()
```
