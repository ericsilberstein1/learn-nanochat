```python
import sys
sys.path.append('../my_nanochat')
import os
import torch
from my_nanochat.my_common import get_base_dir, autodetect_device_type
from my_nanochat.my_checkpoint_manager import build_model
from my_nanochat.my_engine import Engine
```

```python
device_type = autodetect_device_type()
device = torch.device(device_type)
autocast_ctx = torch.amp.autocast(device_type=device_type, dtype=torch.bfloat16) if device_type == "cuda" else nullcontext()
checkpoint_dir = os.path.join(get_base_dir(), "base_checkpoints", "d12")
model, tokenizer, meta_data = build_model(checkpoint_dir, step=21000, device=device, phase="eval")
```

```python
engine = Engine(model, tokenizer)
```

```python
prompt = "The person"
tokens = tokenizer.encode(prompt, prepend=tokenizer.get_bos_token_id())
```

```python
with autocast_ctx:
    samples, _ = engine.generate_batch(tokens, num_samples=10, max_tokens=30)
for sample in samples:
    print(tokenizer.decode(sample) + "\n")
```

```python
def print_10_samples(prompt):
    tokens = tokenizer.encode(prompt, prepend=tokenizer.get_bos_token_id())
    with autocast_ctx:
        samples, _ = engine.generate_batch(tokens, num_samples=10, max_tokens=30, top_k=10)
    for sample in samples:
        print(tokenizer.decode(sample) + "\n")
```

```python
print_10_samples("First take a right on main street")
```
