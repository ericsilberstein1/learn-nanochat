SFT uses ARC-Easy, ARC-Challenge, GSMK8k, "identity", SimpleSpelling and SpellingBee. This notebook shows a few examples of ARC-Easy and ARC-Challenge. For the others see `challenge-26-understand-midtrain/midtrain-data-examples.ipynb`.

```python
import sys
sys.path.append('../my_nanochat')
import os
from my_nanochat.my_common import get_base_dir
from my_tasks.my_arc import MyARC
```

```python
def print_conversation(conversation):
    for message in conversation['messages']:
        print(f"_____{message['role']}_____")
        content = message['content']
        if isinstance(content, str):
            print(content)
        else:
            assert isinstance(content, list)
            for part in content:
                print(f"{part['type']}: {part['text']}")
        print()
```

```python
task = MyARC(subset="ARC-Easy", split="train")
```
